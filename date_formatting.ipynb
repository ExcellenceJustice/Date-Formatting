### Date Conversion

```python
# loading packages

import pandas as pd
```

```python
# importing the file

data = pd.read_excel(r'C:\Users\Justice.Agbadu\Documents\Data_Analytics\Projects\Seun\data_formatting\Nigeria_Datasets_Copy.xlsx', sheet_name='Daily Increase')
print(data)
```

```python
from datetime import datetime
```

```python
L1 = []

for i in range(data['DateNew'].count()):
    L1.append(datetime.strptime(data['DateNew'][i], '%d %B, %Y'))
```

```python
data['DateTime'] = L1
```

```python
data['DateTime']
```

```python
L2 = []

for x in range(data['DateTime'].count()):
    L2.append(data['DateTime'][x].strftime('%m%d%Y'))
```

```python
data['New Date'] = L2
```

```python
data.head()
```

```python
data.drop(['DateTime'], axis=1, inplace=True)
```

```python
data = data.style.hide_index()
```
